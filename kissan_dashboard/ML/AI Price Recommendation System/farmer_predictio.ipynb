```python
import pandas as pd
import joblib
```

```python
model = joblib.load('price_model.pkl')

commodity_encoder = joblib.load('commodity_encoder.pkl')

market_encoder = joblib.load('market_encoder.pkl')

district_encoder = joblib.load('district_encoder.pkl')
```

```python
farmer_crop = 'Wheat'

farmer_market = 'Kotma APMC'

farmer_district = 'Anupur'

farmer_expected_price = 2200
```

```python
sample = pd.DataFrame([{
    'commodity_encoded':commodity_encoder.transform([farmer_crop])[0],
    'market_encoded':market_encoder.transform([farmer_market])[0],
    'district_encoded':district_encoder.transform([farmer_district])[0],
    'Min_Price': 2300,
    'Max_Price': 2450,
    'price_spread': 150,
    'month': 12,
    'week': 52}])
```

```python
predicted_price=model.predict(sample)[0]
print("🌾 Predicted Fair Price:",predicted_price)
```

```python
if farmer_expected_price < predicted_price * 0.9:
    print("⚠️ You may be underpricing your crop.")
elif farmer_expected_price > predicted_price * 1.1:
    print("📈 Your asking price is high.")
else:
    print("✅ Your price looks fair.")
```
